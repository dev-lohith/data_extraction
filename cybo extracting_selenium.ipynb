# Merge all geojsons

```python
# Create an empty GeoDataFrame
india_merged = gpd.GeoDataFrame()

# Directory containing the GeoJSON files
directory_path = '/home/lohith/Documents/git_push/GIS/cybo_polygon_scrap/'  # Replace with the path to your directory

# Loop through the files and merge
for filename in os.listdir(directory_path):
    if filename.endswith('.geojson'):
        file_path = os.path.join(directory_path, filename)
        try:
            gdf = gpd.read_file(file_path)
            india_merged = pd.concat([india_merged, gdf], ignore_index=True)
        except Exception as e:
            print(f"Error reading {file_path}: {str(e)}")

# Save the merged GeoDataFrame to a new GeoJSON file
output_path = './india_merged.geojson'  # Replace with the desired output file path
india_merged.to_file(output_path, driver='GeoJSON')
```

# Final Data

```python
complete_india = gpd.read_file("./india_merged.geojson")
complete_india
```

```python
# Function to count polygons within a geometry
def count_polygons(geometry):
    if geometry.geom_type == 'Polygon':
        return 1
    elif geometry.geom_type == 'MultiPolygon':
        return len(geometry.geoms)
    else:
        return 0

# Count polygons in each GeoDataFrame
count_merged = sum(complete_india['geometry'].apply(count_polygons))

print(f"Total number of polygons in Cybo: {count_merged}")
```

```python
# Simplify the GeoDataFrame to reduce the number of vertices
simplified_india = complete_india.simplify(tolerance=0.01, preserve_topology=True)

# Plot the simplified GeoDataFrame
fig, ax = plt.subplots(figsize=(12, 12))
simplified_india.boundary.plot(ax=ax, linewidth=0.5, color='black')
simplified_india.plot(ax=ax, color='lightblue', alpha=0.7)

# Set axis limits
ax.set_xlim(simplified_india.total_bounds[0], simplified_india.total_bounds[2])
ax.set_ylim(simplified_india.total_bounds[1], simplified_india.total_bounds[3])

# Show the plot
plt.title('Simplified India GeoDataFrame')
plt.show()
```
