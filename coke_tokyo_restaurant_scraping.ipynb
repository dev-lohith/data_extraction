# Restaurant_names

```python
links_refined = pd.read_csv('./links_refined.csv')
```

```python
links_refined_list = links_refined['links'].tolist()
```

```python
type(links_refined_list)
```

```python
output_list = []

for each in links_refined_list:
    response = requests.get(each)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    for link in soup.find_all('span', {"class":"rd-detail-info__rst-name-main"}):
        print(link.get_text())
```

# categories

```python
output_list = []

for each in links_refined_list:
    response = requests.get(each)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    for link in soup.find_all('span',{"property":"v:category"}):
        print(link.get_text())
```

# coordinates_links

```python
soup.find('img').attrs['src']
```

```python
output_list = []

for each in links_refined_list:
    response = requests.get(each)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    output_list.append(soup.find('img').attrs['src'])
```

```python
df = pd.DataFrame (output_list, columns = ['cooordinates'])
```

```python
df.to_csv('./cooordinate_links.csv', index=False)
```
